```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from google.colab import files
import io

# 1. 파일 업로드 및 로드
uploaded = files.upload()
df = pd.read_excel(io.BytesIO(uploaded[list(uploaded.keys())[0]]))

# 2. 상관관계 분석(Pearson)
corr_value = df['studyHour'].corr(df['score'])
print(f"\n상관계수 : {corr_value:.4f}")

# 3. 시각화(산점도 + 회귀선)
plt.figure(figsize=(8, 5))
sns.regplot(x='studyHour', y='score', data=df,line_kws={'color':'red'})
plt.title(f'Correlation : {corr_value:.2f}')
plt.show()
```
