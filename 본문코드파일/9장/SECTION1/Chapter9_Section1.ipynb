```python
# ============================================================
# RMSE / MAE 오차 분석
# 신선식품 유통 기업 판매량 예측 평가
# ============================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error

# ============================================================
# 데이터 생성
# ============================================================

df = pd.DataFrame({
    '날짜': ['2026-06-01', '2026-06-02', '2026-06-03', '2026-06-04', '2026-06-05'],
    '실제 판매량': [100, 120, 150, 110, 130],
    'AI 예측 판매량': [110, 115, 140, 130, 125]
})

# ============================================================
# 오차 계산
# ============================================================

actual = df['실제 판매량']
pred = df['AI 예측 판매량']

mae = mean_absolute_error(actual, pred)
rmse = np.sqrt(mean_squared_error(actual, pred))

# ============================================================
# 결과 출력
# ============================================================

print("=" * 50)
print("예측 오차 분석 결과")
print("=" * 50)
print(f"MAE  : {mae:.2f}")
print(f"RMSE : {rmse:.2f}")
print("=" * 50)

# ============================================================
# 그래프
# ============================================================

plt.figure(figsize=(10, 5))

plt.plot(
    df['날짜'],
    df['실제 판매량'],
    marker='o',
    linewidth=3,
    label='실제 판매량'
)

plt.plot(
    df['날짜'],
    df['AI 예측 판매량'],
    marker='s',
    linewidth=3,
    linestyle='--',
    label='AI 예측 판매량'
)

plt.title('실제 판매량 vs AI 예측 판매량', fontsize=16, fontweight='bold')
plt.xlabel('날짜')
plt.ylabel('판매량')
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.tight_layout()

plt.show()
```
