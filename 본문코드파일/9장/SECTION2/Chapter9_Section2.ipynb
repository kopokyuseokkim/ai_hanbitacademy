```python
# ============================================================
# 상관관계 매트릭스 분석
# 판매량 / 금리 / 소비심리지수
# ============================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# ============================================================
# 데이터 생성
# ============================================================

df = pd.DataFrame({
    '판매량': [100, 110, 105, 95, 90, 85],
    '금리': [1.5, 1.7, 2.0, 2.5, 3.0, 3.5],
    '소비심리지수': [80, 85, 70, 60, 55, 45]
})

# ============================================================
# 상관계수 계산
# ============================================================

corr_matrix = df.corr()

print("=" * 60)
print("상관관계 매트릭스")
print("=" * 60)
print(corr_matrix.round(6))

# ============================================================
# 히트맵 시각화
# ============================================================

plt.figure(figsize=(8, 6))

sns.heatmap(
    corr_matrix,
    annot=True,
    cmap='RdYlBu_r',
    vmin=-1,
    vmax=1,
    linewidths=1,
    fmt='.2f'
)

plt.title(
    '실적,거시경제,심리 지표 상관관계 히트맵',
    fontsize=16,
    fontweight='bold'
)

plt.tight_layout()
plt.show()
```
