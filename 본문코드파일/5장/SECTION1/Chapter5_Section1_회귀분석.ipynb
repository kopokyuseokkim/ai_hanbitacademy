```python
import pandas as pd
import statsmodels.api as sm
import seaborn as sns
import matplotlib.pyplot as plt
from google.colab import files
import io

# 1. 파일 업로드 및 로드
uploaded = files.upload()
df = pd.read_excel(io.BytesIO(uploaded[list(uploaded.keys())[0]]))

# 2. 선형 회귀 분석 (OLS) 수행
X = sm.add_constant(df['studyHour']) #상수항 추가
model = sm.OLS(df['score'], X).fit()
print(model.summary()) # 분석 결과 출력

# 3. 시각화 (산점도 + 회귀선)
sns.regplot(x='studyHour', y='score', data=df, line_kws={'color':'red'})
plt.show()
```
